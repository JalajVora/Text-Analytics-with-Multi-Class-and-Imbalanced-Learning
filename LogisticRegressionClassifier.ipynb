```python
from corpus import load_meta_data
from features import load_features
```

```python
meta_data = load_meta_data()
target = meta_data.guten_genre
meta_data
```

```python
features = load_features().dropna()
del features['book_id']
features
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
```

```python
features_scal = pd.DataFrame(
    StandardScaler().fit_transform(features),
    index=features.index,
    columns=features.columns
  )

features_scal
```

```python
X_train, X_test = train_test_split(features_scal, test_size = 0.33, random_state = 0, stratify=target[features_scal.index])
y_train = target[X_train.index]
y_test = target[X_test.index]
```

```python
sampler = RandomOverSampler(random_state=0)
X_train_samp, y_train_samp = sampler.fit_resample(X_train, y_train)
```

```python
model = LogisticRegression()
model.fit(X_train_samp, y_train_samp)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
mat = confusion_matrix(y_test, y_pred)
sns.heatmap(mat, annot=True, fmt = '.4g')
plt.xlabel('predicted label')
plt.ylabel('true label')
print('Score on training set: ', model.score(X_train_samp, y_train_samp))
print('Score on test set:     ', model.score(X_test, y_test ))
plt.show()
```
